```r
data <- read.csv("heart.csv")
```

```r
count<- 0
for (i in 1:nrow(data)){
    count <- count +1
    }
print(paste("Number of records :", count))
```

```r
min_val <- data$RestingBP[1]
max_val <- data$RestingBP[1]
sum_val <- 0
for (i in 1:nrow(data)){
    if(data$RestingBP[i] < min_val){
        min_val <- data$RestingBP[i]
    }
    if(data$RestingBP[i] > max_val){
        max_val <- data$RestingBP[i]
     }
    sum_val <- sum_val + data$RestingBP[i]
}

mean_val <- sum_val / nrow(data)

summary <- paste("Minimum value :", min_val, "Maximum value :", max_val, "Mean Value :", mean_val)
print(summary)
```

```r
count <- 0
for (i in 1:nrow(data)){
    if(data$RestingBP[i] > mean_val){
        count <- count + 1
    }
}
print (paste("Patients with RestingBP Greater than mean :", count))
```

```r
unique_types <- c()
for (i in 1:nrow(data)){
    value <- data$ChestPainType[i]
    if (!(value %in% unique_types)){
        unique_types <- c(unique_types, value)
    }
}
print("Unique Chest Pain Type :")
print (unique_types)
```

```r
count <- 0
for (i in 1:nrow(data)) {
  if (data$Sex[i] == "Female" && data$RestingBP[i] > 140 && count < 10) {
    print(paste(data$Age[i]))
    count <- count + 1
  }
}
```

```r
check_cholesterol <- function(row) {
  chol_level <- ifelse(row$Cholesterol < 200, "Low", "High")
  return(paste("Age:", row$Age, " | Sex:", row$Sex, " | Cholesterol:", chol_level))
}

for (i in 1:nrow(data)) {
  row <- data[i, ]
  print(check_cholesterol(row))
}
```
